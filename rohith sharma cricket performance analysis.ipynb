```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to plot Year vs Balls for ODI
def runs_odi(data):
    plt.figure(figsize=(8,7))
    plt.plot(data['Runs'], data['Year'], color='black', marker='*',linewidth='3',markersize='10')
    plt.title('Rohit Sharma ODI performance (Runs vs Years)')
    plt.xlabel('Runs')
    plt.ylabel('Year')
    plt.show()

# Function to plot Year vs Strike Rate for ODI
def srate_odi(data):
    plt.figure(figsize=(8,7))
    plt.plot(data['SR'], data['Year'], color='black', marker='*',linewidth='3',markersize='10')
    plt.title('Rohit Sharma ODI performance (Strike Rate vs Years)')
    plt.xlabel('Strike Rate')
    plt.ylabel('Year')
    plt.show()

# Function to plot Innings vs Years for ODI
def innings_odi(data):
    plt.figure(figsize=(8,7))
    plt.plot(data['Innings'], data['Year'], color='black', marker='*',linewidth='3',markersize='10')
    plt.title('Rohit Sharma ODI performance (Innings vs Years)')
    plt.xlabel('Innings')
    plt.ylabel('Year')
    plt.show()

# Function to plot Year vs Runs for T20
def runs_t20(data):
    plt.figure(figsize=(8,7))
    plt.plot(data['Runs'], data['Year'], color='black', marker='*',linewidth='3',markersize='10')
    plt.title('Rohit Sharma T20 performance (Runs vs Years)')
    plt.xlabel('Runs')
    plt.ylabel('Year')
    plt.show()

# Function to plot Year vs Strike Rate for T20
def srate_t20(data):
    plt.figure(figsize=(8,7))
    plt.plot(data['SR'], data['Year'], color='black', marker='*',linewidth='3',markersize='10')
    plt.title('Rohit Sharma T20 performance (Strike Rate vs Years)')
    plt.xlabel('Strike Rate')
    plt.ylabel('Year')
    plt.show()

# Function to plot Innings vs Years for T20
def innings_t20(data):
    plt.figure(figsize=(8,7))
    plt.plot(data['Innings'], data['Year'], color='black', marker='*',linewidth='3',markersize='10')
    plt.title('Rohit Sharma T20 performance (Innings vs Years)')
    plt.xlabel('Innings')
    plt.ylabel('Year')
    plt.show()

# Function to plot Year vs Runs for test
def runs_test(data):
    plt.figure(figsize=(8,7))
    plt.plot(data['Runs'], data['Year'], color='black', marker='*',linewidth='3',markersize='10')
    plt.title('Rohit Sharma TEST performance (Runs vs Years)')
    plt.xlabel('Runs')
    plt.ylabel('Year')
    plt.show()

# Function to plot Year vs Strike Rate for test
def srate_test(data):
    plt.figure(figsize=(8,7))
    plt.plot(data['SR'], data['Year'], color='black', marker='*',linewidth='3',markersize='10')
    plt.title('Rohit Sharma TEST performance (Strike Rate vs Years)')
    plt.xlabel('Strike Rate')
    plt.ylabel('Year')
    plt.show()

# Function to plot Innings vs Years for test
def innings_test(data):
    plt.figure(figsize=(8,7))
    plt.plot(data['Innings'], data['Year'], color='black', marker='*',linewidth='3',markersize='10')
    plt.title('Rohit Sharma TEST performance (Innings vs Years)')
    plt.xlabel('Innings')
    plt.ylabel('Year')
    plt.show()

def main():
    format_choice = input("Enter 'ODI' or 'T20' or 'TEST' to choose the format: ").upper()

    if format_choice == 'ODI':
        odi_data = pd.read_csv('/content/drive/MyDrive/archive (3)/ODI.csv')
        choice = int(input("Enter:\n1 for Runs vs Years\n2 for Strike Rate vs Years\n3 for Innings vs Years:\n "))

        if choice == 1:
            runs_odi(odi_data)
        elif choice == 2:
            srate_odi(odi_data)
        elif choice == 3:
            innings_odi(odi_data)
        else:
            print("Invalid choice. Please enter correct choice.")

    elif format_choice == 'T20':
        t20_data = pd.read_csv('/content/drive/MyDrive/archive (3)/T20.csv')
        choice = int(input("Enter:\n1 for Runs vs Years\n2 for Strike Rate vs Years\n3 for Innings vs Years:\n "))

        if choice == 1:
            runs_t20(t20_data)
        elif choice == 2:
            srate_t20(t20_data)
        elif choice == 3:
            innings_t20(t20_data)
        else:
            print("Invalid choice. Please enter correct choice.")
    elif format_choice == 'TEST':
        odi_data = pd.read_csv('/content/drive/MyDrive/archive (3)/TEST.csv')
        choice = int(input("Enter:\n1 for Runs vs Years\n2 for Strike Rate vs Years\n3 for Innings vs Years:\n "))

        if choice == 1:
            runs_test(odi_data)
        elif choice == 2:
            srate_test(odi_data)
        elif choice == 3:
            innings_test(odi_data)
        else:
            print("Invalid choice. Please enter correct choice.")

    else:
        print("Invalid format choice. Please enter either 'ODI' or 'T20'.")

if __name__ == "__main__":
    main()
```
